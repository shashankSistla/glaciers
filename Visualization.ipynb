```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np

file_path = 'output/demo/01_prepare_dem/output/G007276E45599N_dem.tif'

with rasterio.open(file_path) as src:
    band1 = src.read(1)
    nodata_value = src.nodata

    # Mask 'no data' values if a nodata value is set
    if nodata_value is not None:
        band1 = np.ma.masked_where(band1 == nodata_value, band1)

    plt.imshow(band1, cmap='viridis')  # Choose a colormap as needed
    plt.colorbar(label='Value')
    plt.axis('off')
    plt.show()
```
